```python
import pandas as pd
```

```python
df1 = pd.read_csv('data2.csv', encoding='cp949')
df1
```

```python
df1.columns
```

```python
df1.drop([0, 1], axis = 0, inplace=True)
```

```python
df1.drop(columns=['순번', '상품코드', ' 합계'], inplace=True)
```

```python
df2 = pd.DataFrame(df1.set_index('상품명').stack().reset_index())
df2
```

```python
df3 = df2.rename(columns={'level_1':'지점', 0:'판매량'})
df3
```

```python
df3['상품명'].unique()
```

```python
df3.info()
```

```python
df3['판매량'] = df3['판매량'].str.replace(',', '').astype(int)
```

```python
df3['판매량']
```

```python
def func_type(row):
    if 'BW1' in row:
        return 'B type'
    if 'CAT' in row:
        return 'C type'
    if 'G1' in row:
        return 'G type'
    if 'STL' in row:
        return 'S type'
    else:
        return 'nomal type'
```

```python
df3['상품타입'] = df3['상품명'].apply(func_type)
```

```python
df3
```

```python
df3['상품타입'].unique()
```

```python
df4=pd.pivot_table(df3, index=['지점', '상품타입'], 
                          values = '판매량', aggfunc = 'sum').reset_index()
df4
```

```python
import matplotlib.pyplot as plt
plt.rc('font',family='Malgun Gothic')
```

```python
idx_max = df4.loc[df4.groupby('상품타입')['판매량'].idxmax()]
idx_max
```

```python
df5 = df4.sort_values(by = '판매량', ascending=False).head(200)
df5
```

```python
df5.to_csv('result_0625.csv', encoding = 'cp949')
```
