```python
import pandas as pd
```

```python
df1 = pd.read_excel('data3.xls')
df1
```

```python
df1.columns
```

```python
df1.info()
```

```python
df1.drop(columns=['고객명', '이름'], inplace=True)
```

```python
df1['주문연도'] = df1['주문일'].dt.year
df1['주문월'] = df1['주문일'].dt.month
df1['주문일자'] = df1['주문일'].dt.day
```

```python
df1
```

```python
df1.info()
```

```python
df1['고객분류'].unique()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.rcParams['font.family'] = 'Malgun Gothic'  
plt.rcParams['font.size'] = 10 
plt.rcParams['figure.figsize'] = (15, 10)
plt.rcParams['axes.unicode_minus'] = False  
```

```python
year_Sales = df1.groupby(['주문연도', '고객분류'])['이익'].sum().reset_index()
sns.lineplot(data=year_Sales, x='주문연도', y='이익', hue='고객분류')
```

```python
year_product = df1.groupby(['주문연도', '제품대분류'])['이익'].sum().reset_index()
sns.lineplot(data=year_product, x='주문연도', y='이익', hue='제품대분류')
```

```python
sns.scatterplot(data=df1, x='매출', y='이익')
plt.title('테이블 매출과 이익의 관계')
plt.show()
```

```python
rising_profit=df1.loc[
    (df1['고객분류'].isin(['개인고객','홈오피스']))&
    (df1['제품대분류'].isin(["전자제품","사무용품"]))]
```

```python
rising_profit_time = rising_profit.groupby(['제품소분류', '주문연도'])['이익'].sum().reset_index()
sns.lineplot(rising_profit_time, x='주문연도', y='이익', 
             hue='제품소분류', markers=True, style='제품소분류')
plt.title('(개인고객&홈오피스) 제품별 이익')
plt.show()
```

```python
df1.columns
```

```python
Sub_class = df1.groupby(['제품소분류'])['이익'].sum().reset_index()
plt.title('제품별 총 이익')
sns.barplot(data=Sub_class, x='제품소분류', y='이익')
plt.xticks(rotation=45)
plt.show()
```

```python
state_profit = df1.groupby(['도(State)'])['이익'].mean().reset_index()
sns.barplot(data=state_profit, x='도(State)', y='이익')
plt.title('도(State)별 평균 이익')
plt.xticks(rotation=45)
plt.show()
```

```python
state_profit = df1.groupby(['도(State)', '배송방식'])['이익'].sum().reset_index()
sns.barplot(data=state_profit, x='도(State)', y='이익', hue='배송방식')
plt.title('도(State)별 배송방식에 따른 총이익')
plt.legend(title='배송방식')
plt.xticks(rotation=45)
plt.show()
```

```python
df2=df1.loc[df1['제품소분류'].str.contains("테이블")]
```

```python
year_Profit = df2.groupby(['지역', '주문연도'])['이익'].sum().reset_index()
sns.barplot(data=year_Profit, x='지역', y='이익', hue='주문연도')
plt.title("지역별 테이블 이익")
plt.show()
```

```python
sns.scatterplot(data=df2, x='할인율', y='이익')
plt.title('테이블 할인율과 이익의 관계')
plt.show()
```

```python
sns.scatterplot(data=df2, x='매출', y='이익')
plt.title('테이블 매출과 이익의 관계')
plt.show()
```

```python
year_Profit = df2.groupby(['고객분류', '주문연도'])['이익'].sum().reset_index()
sns.barplot(data=year_Profit, x='고객분류', y='이익', hue='주문연도')
plt.title("고객별 테이블 이익")
plt.show()
```

```python
df3=df2.loc[df2['이익']>0]
```

```python
year_Profit_plus = df3.groupby(['지역', '주문연도'])['이익'].sum().reset_index()
sns.barplot(data=year_Profit_plus, x='지역', y='이익', hue='주문연도')
plt.title("지역별 테이블 플러스 이익")
plt.show()
```

```python
df4 = df2.loc[(df2['주문연도']==2013)&(df2['지역'].str.contains("수도권"))]
```

```python
city_profit = df4.groupby('시군구(City)')['이익'].sum().reset_index()
sns.barplot(data=city_profit, x='시군구(City)', y='이익')
plt.title('시군구별 테이블 이익 (2013년, 수도권)')
plt.show()
```
