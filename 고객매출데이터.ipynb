# 액세서리 위주로 시계열 분석

```python
df_acc = df1.loc[df1['제품소분류'] == '액세서리']
df_acc
```

```python
annual_profit = pd.pivot_table(data=df_acc, index='주문일', values = '이익', 
                               aggfunc='sum').reset_index()
sns.lineplot(annual_profit, x='주문일', y='이익')
plt.show()
```

```python
import statsmodels.tsa.api as tsa
```

```python
df_acc_series = pd.pivot_table(data=df_acc, index='주문일', values='이익', aggfunc='sum')
```

```python
df_acc_series
```

```python
y = df_acc_series['이익'].resample('1W').sum()
```

```python
y
```

```python
y.isnull().sum()
```

```python
y1 = y
```

```python
y1
```

```python
model_series = tsa.seasonal_decompose(y1, model='additive')
fig = model_series.plot()
plt.show()
```

```python
import itertools 
```

```python
p = d = q = range(0, 2)
pdq = list(itertools.product(p, d, q))
seasonal_pdq = [(x[0], x[1], x[2], 12) for x in list(itertools.product(p, d, q))]
```

```python
seasonal_pdq
```

```python
print('SARIMAX: {} x {}'.format(pdq[1], seasonal_pdq[1]))
print('SARIMAX: {} x {}'.format(pdq[1], seasonal_pdq[2]))
print('SARIMAX: {} x {}'.format(pdq[2], seasonal_pdq[3]))
print('SARIMAX: {} x {}'.format(pdq[2], seasonal_pdq[4]))
```

```python
param_list = []
param_seasonal_list = []
results_AIC_list = []
```

```python
for param in pdq:
     for param_seasonal in seasonal_pdq:
        try:
            mod = tsa.statespace.SARIMAX(y1,order=param,
                                            seasonal_order=param_seasonal,
                                            enforce_stationarity=False,
                                            enforce_invertibility=False)
            results = mod.fit()
            param_list.append(param)
            param_seasonal_list.append(param_seasonal)
            results_AIC_list.append(results.aic)
        except:
             continue
```

```python
ARIMA_list = pd.DataFrame({'Parameter':param_list,'Seasonal':param_seasonal_list,'AIC':results_AIC_list})
ARIMA_list.to_excel('arima_model_list.xlsx')
```

```python
ARIMA_list.sort_values(by='AIC')
```

```python
mod = tsa.statespace.SARIMAX(y1,order=(0, 1, 1),seasonal_order=(0, 1, 1, 12),
                              enforce_stationarity=False, enforce_invertibility=False)
results = mod.fit()
print(results.summary())
```

```python
results.plot_diagnostics(figsize=(16, 8))
plt.show()
```

```python
results.get_prediction()
```

```python
y1.head()
```

```python
pred = results.get_prediction(start=pd.to_datetime('2012-01-15'), dynamic=False)
pred_ci = pred.conf_int()

ax = y1.plot(label='observed')
pred.predicted_mean.plot(ax=ax, label='One-step ahead Forecast', alpha=.7, figsize=(14, 7))
ax.fill_between(pred_ci.index,pred_ci.iloc[:, 0],pred_ci.iloc[:, 1], color='k', alpha=.2)

plt.legend()
plt.show()
```

```python
pred = results.get_prediction(start=pd.to_datetime('2012-01-15'), dynamic=False)
pd.DataFrame(pred.predicted_mean).reset_index()
```

```python
y_forecasted = pred.predicted_mean
y_truth = y['2012-01-01':]
mse = ((y_forecasted - y_truth) ** 2).mean()
print('MSE {}'.format(round(mse, 2)))
```

```python
pred_uc = results.get_forecast(steps=50)
pred_ci = pred_uc.conf_int() 
ax = y1.plot(label='observed', figsize=(14, 7))
pred_uc.predicted_mean.plot(ax=ax, label='Forecast')
ax.fill_between(pred_ci.index,
                 pred_ci.iloc[:, 0],
                 pred_ci.iloc[:, 1], color='k', alpha=.25)
plt.title('액세서리 이익 예측')
plt.legend()
plt.show()
```
