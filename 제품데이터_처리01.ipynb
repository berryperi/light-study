```python
df1 = pd.read_excel('data1.xlsx', skiprows = 1)
df1 = df1.drop(columns=['Unnamed: 7', '판매'])
df1
```

```python
df2 = pd.DataFrame(df1.iloc[ :, 6:-2].set_index(['제품명']).stack()).reset_index()
df3 = df2.rename(columns={'level_1':'날짜', 0:'재고량'})
```

```python
df1.columns
```

```python
df4 = df1[['구분', 0, '카테고리명', '자재그룹', '자재그룹명', '제품코드', '제품명', '안전재고', ' 분류']]
df4
```

```python
df5 = pd.merge(df4, df3, on='제품명', how='left')
```

```python
df5
```

```python
df5.info()
```

```python
df5['공급월'] = df5['날짜'].dt.month
df5['공급일'] = df5['날짜'].dt.day
df5['공급연도'] = df5['날짜'].dt.year
df5['공급주'] = df5['날짜'].dt.week
```
